```python
#Create a generalized function for speech
import pyttsx3

def speak(text):
    #initialize the class
    engine = pyttsx3.init()

    #change the rate of the speech
    engine.setProperty('rate' , 110)

    #change the voice of the speaker
    #0:male , 1:female
    voices = engine.getProperty('voices')
    engine.setProperty('voice' , voices[0].id)

    #utter the word
    engine.say(text)

    engine.runAndWait()
    
```

```python
# libraries for Web Scrapping
import requests
from bs4 import BeautifulSoup
import urllib.request

#libraries for Geo-loaction
from geopy.geocoders import Nominatim

#Taking the input of a city
address = input("Enter the city = ")

#Latitude & Longitude
geo_location = Nominatim(user_agent = 'Dipanwita')
location = geo_location.geocode(address)
lat = location.latitude
long = location.longitude

#page requests
page = requests.get('https://weather.com/en-IN/weather/today/l/{},{}?par=google&temp=c'.format(lat , long))

#soup
soup = BeautifulSoup(page.text , 'lxml')

#Location
location = soup.find('h1', class_='CurrentConditions--location--1Ayv3')
speak(location.text)
print(location.text)

#Temperature in degree
temp = soup.find('span' , class_ = 'CurrentConditions--tempValue--3KcTQ')
speak('Temperature is'+ temp.text.split('°')[0] +' degrees')
print('Temperature is '+ temp.text.split('°')[0] +' degrees')

#Weather Condition
weather = soup.find('div' , class_='CurrentConditions--phraseValue--2xXSr')
speak('Weather is' + weather.text)
print('Weather is ' + weather.text)

try:
    condition = soup.find('div' , class_='CurrentConditions--precipValue--RBVJT')
    speak(condition.text)
    print(condition.text)
except:
    speak("Prediction not avialable !!")
    print("Prediction not avialable !!")

#Air quality
try:
    air_qual_index = soup.find('text' , class_='DonutChart--innerValue--k2Z7I')
    speak('Air quality index is' + air_qual_index.text)
    print('Air quality index is ' + air_qual_index.text)
except:
    speak("Prediction not avialable !!")
    print("Prediction not avialable !!")  
    
try:
    air_qual = soup.find('span' , class_='AirQualityText--severity--1VMr2')
    speak('It' + air_qual.text)
    print('It ' + air_qual.text)
except:
    speak("Prediction not avialable !!")
    print("Prediction not avialable !!")

#Humidity    
try:   
    humidity = soup.find('span' , attrs = {'data-testid':"PercentageValue"} )
    speak(humidity.text + 'Humidity')
    print(humidity.text + ' Humidity')
        
except:
    speak("Prediction not avialable !!")
    print("Prediction not avialable !!")

#Wind Speed    
try:  
    wind = soup.find('span' , class_ = 'Wind--windWrapper--1Va1P undefined' )
    speak('Wind' + wind.text)
    print('Wind ' + wind.text)
        
except:
    speak("Prediction not avialable !!")
    print("Prediction not avialable !!")
    
    
speak("Thank you ")  
print("Thank you ") 
```
